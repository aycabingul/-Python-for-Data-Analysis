## **8.1 Hierarchical Indexing**

```python
import pandas as pd
import numpy as np
data = pd.Series(np.random.randn(9),
             index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
                    [1, 2, 3, 1, 3, 1, 2, 2, 3]])
```

```python
data
```

```python
data.index
```

```python
data["b"]
```

```python
data["b" : "c"]
```

```python
data.loc[["b","d"]]
```

```python
data.loc[:,2]
```

```python
data.unstack()
```

```python
data.unstack().stack()
```

### **Reordering and Sorting Levels**

```python
frame.swaplevel('key1','key2')
```

```python
frame.sort_index(level=1)
```

```python
frame.swaplevel(0,1).sort_index(level=0)
```

### **Summary Statistics by Level**

```python
frame.sum(level='key2')
```

```python
frame.sum(level='color',axis=1)
```

## **8.2 Combining and Merging Datasets**

### **Database-Style DataFrame Joins**

```python
df1=pd.DataFrame({'data1':range(7),
                'key':['b','b','a','c','a','a','b']})
```

```python
df2=pd.DataFrame({  'data2':range(3),'key':['a','b','d']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
df3 = pd.DataFrame({'data1': range(7),'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b']})
```

```python
df4 = pd.DataFrame({'data2': range(3),'rkey': ['a', 'b', 'd']})
```

```python
pd.merge(df3,df4,left_on='lkey',right_on='rkey')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
df1 = pd.DataFrame({'data1': range(6),'key': ['b', 'b', 'a', 'c', 'a', 'b']})
```

```python
df2 = pd.DataFrame({'data2': range(5),'key': ['a', 'b', 'a', 'b', 'd']})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,how='inner')
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
```

```python
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```
